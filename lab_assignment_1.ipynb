```python
# Graph created based on CS Academy editor data

graph = {
    0: [1, 2, 4],
    1: [0, 3, 4],
    2: [0, 3, 5],
    3: [1, 2, 5],
    4: [0, 1, 5],
    5: [2, 3, 4]
}

print("Graph Created Successfully!")
```

```python
# BFS Implementation

def bfs(graph, start):
    visited = []          # To store visited nodes
    queue = [start]       # Start from given node

    while queue:
        node = queue.pop(0)   # Take first element from queue
        if node not in visited:
            visited.append(node)
            queue.extend(graph[node])  # Add its neighbors
    return visited

# Run BFS from node 0
print("BFS Traversal:", bfs(graph, 0))
```

```python
# DFS Implementation

def dfs(graph, start, visited=None):
    if visited is None:
        visited = []
    visited.append(start)
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)
    return visited

# Run DFS from node 0
print("DFS Traversal:", dfs(graph, 0))
```

### Explanation:
- BFS explores all nodes level by level using a queue.
- DFS explores as deep as possible before backtracking, using recursion.

```python
import time
from collections import deque
```

```python
# Function to get all possible next states from current state
def get_next_states(state):
    x, y = state
    jug1, jug2 = 4, 3  # capacities
    next_states = []

    # Fill jug1
    next_states.append((jug1, y))
    # Fill jug2
    next_states.append((x, jug2))
    # Empty jug1
    next_states.append((0, y))
    # Empty jug2
    next_states.append((x, 0))
    # Pour jug1  jug2
    pour = min(x, jug2 - y)
    next_states.append((x - pour, y + pour))
    # Pour jug2  jug1
    pour = min(y, jug1 - x)
    next_states.append((x + pour, y - pour))

    return next_states
```

```python
def bfs(start, goal):
    start_time = time.time()
    visited = set()
    queue = deque([(start, [start])])

    while queue:
        (x, y), path = queue.popleft()
        if (x, y) == goal:
            end_time = time.time()
            return path, end_time - start_time, len(visited)

        visited.add((x, y))
        for next_state in get_next_states((x, y)):
            if next_state not in visited:
                queue.append((next_state, path + [next_state]))

    return None, None, None
```

```python
def dfs(start, goal):
    start_time = time.time()
    visited = set()
    stack = [(start, [start])]

    while stack:
        (x, y), path = stack.pop()
        if (x, y) == goal:
            end_time = time.time()
            return path, end_time - start_time, len(visited)

        visited.add((x, y))
        for next_state in get_next_states((x, y)):
            if next_state not in visited:
                stack.append((next_state, path + [next_state]))

    return None, None, None
```

```python
start = (0, 0)   # both jugs empty
goal = (2, 0)    # we need 2 liters in jug1

# Run BFS
bfs_path, bfs_time, bfs_steps = bfs(start, goal)
print("BFS Solution Path:", bfs_path)
print("BFS Steps Taken:", len(bfs_path) - 1)
print("BFS Time:", bfs_time)
print("BFS States Explored:", bfs_steps)

print("\n----------------------------------\n")

# Run DFS
dfs_path, dfs_time, dfs_steps = dfs(start, goal)
print("DFS Solution Path:", dfs_path)
print("DFS Steps Taken:", len(dfs_path) - 1)
print("DFS Time:", dfs_time)
print("DFS States Explored:", dfs_steps)
```

```python
import heapq  # for priority queue

# Define Maze: S = Start, G = Goal, # = Wall, . = Empty
maze = [
    ['S', '.', '.', '#', '.'],
    ['.', '#', '.', '#', '.'],
    ['.', '#', '.', '.', '.'],
    ['.', '#', '#', '.', '.'],
    ['.', '.', '.', '#', 'G']
]

# Manhattan distance heuristic
def heuristic(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

# A* Algorithm
def astar(maze, start, goal):
    rows, cols = len(maze), len(maze[0])
    open_list = []
    heapq.heappush(open_list, (0, start))
    came_from = {}
    g_score = {start: 0}

    while open_list:
        _, current = heapq.heappop(open_list)

        if current == goal:
            # Reconstruct path
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            path.reverse()
            return path

        # Explore neighbors
        for dx, dy in [(-1,0),(1,0),(0,-1),(0,1)]:
            neighbor = (current[0]+dx, current[1]+dy)
            if 0 <= neighbor[0] < rows and 0 <= neighbor[1] < cols:
                if maze[neighbor[0]][neighbor[1]] == '#':
                    continue  # skip walls

                tentative_g = g_score[current] + 1
                if neighbor not in g_score or tentative_g < g_score[neighbor]:
                    g_score[neighbor] = tentative_g
                    f_score = tentative_g + heuristic(neighbor, goal)
                    heapq.heappush(open_list, (f_score, neighbor))
                    came_from[neighbor] = current

    return None  # no path found

# Run A*
start = (0, 0)
goal = (4, 4)
path = astar(maze, start, goal)

print("A* Path Found:", path)

# Visualize the path
for (x, y) in path:
    if maze[x][y] not in ['S', 'G']:
        maze[x][y] = '*'

print("\nMaze with Path:\n")
for row in maze:
    print(' '.join(row))

```
